```python
import cloudscraper
import pandas as pd
from io import StringIO
import time
import random
```

```python
years = list(range(2024, 2017, -1))
```

```python
scraper = cloudscraper.create_scraper()
```

```python
table_urls = [f"https://www.transfermarkt.de/bundesliga/startseite/wettbewerb/L1/plus/?saison_id={year}" for year in years]
```

```python
table_urls
```

```python
data = scraper.get(table_urls[2])
```

```python
data.text
```

```python
html_data = StringIO(data.text)
table = pd.read_html(html_data)[1]
```

```python
table = table.drop(table.index[-1])
```

```python
table.columns
```

```python
table.rename(columns={"Kader": "mean_age", "ø-Marktwert": "mean_value", "Verein.1": "team"}, inplace=True)
table = table[["team", "mean_age", "mean_value"]]
table["mean_age"] = table["mean_age"] / 10
table['mean_value'] = table['mean_value'].str.extract(r'([\d,]+)')
table['mean_value'] = table['mean_value'].str.replace(',', '.').astype(float)
table
```

```python
for year in years:
    table_url = f"https://www.transfermarkt.de/bundesliga/startseite/wettbewerb/L1/plus/?saison_id={year}"
    data = scraper.get(table_url)
    time.sleep(random.uniform(4,8))
    html_text = StringIO(data.text)
    table = pd.read_html(html_text)[1]
    table = table.drop(table.index[-1])
    table.rename(columns={"Kader": "mean_age", "ø-Marktwert": "mean_value", "Verein.1": "team"}, inplace=True)
    table["mean_age"] = table["mean_age"] / 10
    table = table[["team", "mean_age", "mean_value"]]
    table['mean_value'] = table['mean_value'].str.extract(r'([\d,]+)')
    table['mean_value'] = table['mean_value'].str.replace(',', '.').astype(float)
    table["season"] = year + 1
    table.to_csv(f"team_values/values_{year + 1}.csv")
```
